```python
path = 'emails_cleaned.csv'
text_df = pd.read_csv(path)
```

```python
text_senderRecieverTextdf =text_df[['X-From','X-To','content']]
```

Preoprocessing

```python
#remove digiu
text_senderRecieverTextdf['content'] = text_senderRecieverTextdf['content'].str.replace(r'\d+', '', regex=True)
```

```python
# NLP text removal package
import pandas as pd
import neattext.functions as nfx
```

```python
# text type transformation
text_senderRecieverTextdf['content'] =text_senderRecieverTextdf['content'] .astype(str)
```

```python
# Clean Text:stopwords,special charac
text_senderRecieverTextdf['content'] = text_senderRecieverTextdf['content'].apply(nfx.remove_stopwords)
# Clean Text:stopwords,special charac
text_senderRecieverTextdf['content'] = text_senderRecieverTextdf['content'].apply(nfx.remove_special_characters)
```

```python
text_senderRecieverTextdf
```

Compare

```python
cluster_max_sepate_df
```

```python
max_weight_row = cluster_max_sepate_df[cluster_max_sepate_df['Number of Information Exchange'] == cluster_max_sepate_df['Number of Information Exchange'].max()]
```

```python
sender_fromMax = max_weight_row['Sender'].tolist()[0]
receiver_fromMax = max_weight_row['Receiver'].tolist()[0]
```

```python
#query for the text in 
maxFlowEDGEDF = text_senderRecieverTextdf[(text_senderRecieverTextdf['X-From'] == sender_fromMax) & 
                                      (text_senderRecieverTextdf['X-To'] == receiver_fromMax)]
```

```python
maxFlowEDGEDF
```

```python
max_edge_simiality_list = maxFlowEDGEDF['content'].tolist()
```

```python
max_edge_simiality_list
```

```python
# Construct another list
senderName  ='Kate Symes'
receiverName = 'Mark Guzman'
```

```python
#query for the text in 
anotherEdge = text_senderRecieverTextdf[(text_senderRecieverTextdf['X-From'] == senderName) & 
                                      (text_senderRecieverTextdf['X-To'] == receiverName)]
```

```python
anotheredge_simiality_list = anotherEdge['content'].tolist() 
```

```python


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
list_1_compared = max_edge_simiality_list
list_2_compared = anotheredge_simiality_list
combined_texts = list_1_compared + list_2_compared
tf_vectorizer = TfidfVectorizer()
tfidf_matrix = tf_vectorizer.fit_transform(combined_texts)
cosine_sim_matrix = cosine_similarity(tfidf_matrix[:len(list_1_compared)], tfidf_matrix[len(list_1_compared):])
average_similarity = np.mean(cosine_sim_matrix)
print("Average Cosine Similarity:", average_similarity)
```
